```python
import os
import re

# List all files in the current directory
all_files = os.listdir('.')
print("Files in the current directory:")
print(all_files)

# Define a regular expression pattern to match files with (number)
# '\(' matches a literal opening parenthesis
# '\d+' matches one or more digits
# '\)' matches a literal closing parenthesis
pattern = r'\(\d+\)'

# Filter files that match the pattern
files_to_delete = [f for f in all_files if re.search(pattern, f)]

print("\nFiles to delete:")
print(files_to_delete)

# Delete the identified files
for file_name in files_to_delete:
    try:
        os.remove(file_name)
        print(f"File '{file_name}' deleted successfully.")
    except Exception as e:
        print(f"An error occurred while deleting '{file_name}': {e}")
```
